### Gathering data from Internet

```python
#url of tweet image predictions.
url = 'https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv'

#gathering the tweet image predictions programmatically using os and requests.
folder_name = 'files'
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

response = requests.get(url)

with open(os.path.join(folder_name, url.split('/')[-1]), mode = 'wb') as file:
        file.write(response.content)
```

```python
#read in the predictions file.
df_preds = pd.read_csv('files/image-predictions.tsv', sep='\t')
```

### Gathering data using an API

```python
#Tweepy authentication information.
consumer_key = ''
consumer_secret = ''
access_token = ''
access_secret = ''

#Creating variable for api query.
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, wait_on_rate_limit = True, wait_on_rate_limit_notify = True)
```

```python
# List of tweet_ids from archive df
id_list = list(df_archive.tweet_id)

#Creating a text file dump of status information for all the tweet id's in the twitter archive.
with open('tweets_data_json.txt', 'a', encoding = 'utf8') as file:
    for tweet_id in id_list:
        try:
            tweet = api.get_status(tweet_id, tweet_mode='extended')
            json.dump(tweet._json, file)
            file.write('\n')
        except:
            print("Encountered error on id {}".format(tweet_id))
```

```python
#Creating a list for organising dump data
dump_data = []

tweet_dump = open('tweets_data_json.txt', 'r')

#For each line (tweet) in dump data, append to the list
for line in tweet_dump:
    try:
        tweet = json.loads(line)
        dump_data.append(tweet)
    except:
        contine
        
tweet_dump.close()
```

```python
df_info = pd.DataFrame()
```

```python
# Add selected variables to tweet_info DataFrame
df_info['tweet_id'] = list(map(lambda tweet: tweet['id'], dump_data))
df_info['retweets'] = list(map(lambda tweet: tweet['retweet_count'], dump_data))
df_info['likes'] = list(map(lambda tweet: tweet['favorite_count'], dump_data))
```

### Quality
##### Archive Table
- Some rows are retweets and will need to be removed.
- Text variable is truncated and ideally all content should be visible.
- HTML character references for ampersand are visible in the text variable and should be replaced with '&'.
- Source entries are messy HTML anchor tags and should be cleaned and turned into a categorical variable.
- Names of dogs are missing or are incorrect. Names should be corrected where possible or replaced with NaN.
- Entry has the denominator of 0 and will need to be changed to the correct denominator displayed in text.
- Dogs stages listed as 'None' will need to be changed to NaN.
- Irrelevant columns for this analysis need to be removed.
- Some numerators in the text section are listed as floats and this has led to inconsistencies in the numerator variable. Numerators and denominators will need to be changed to floats and numerator values changed to correspond with those in the text variable.

##### Predictions Table
- Underscores are evident between the words in the p1, p2 and p3 variables and will need to be replaced with spaces.
- Capitalisation inconsistency in p1, p2 and p3 columns.


### Tidyness

##### Archive Table
- Dog stage variable is spread across 4 columns. Needs to be consolidated to one.
- Data and time variables are joined in a single column and are not datetime objects

##### All tables
- Tweet id variable is an integer and will need to be changed to a string.
- Datasets need to be joined

## Cleaning

```python
# Creating copies of above dataframes for use in cleaning
archive_clean = df_archive.copy()
pred_clean = df_preds.copy()
info_clean = df_info.copy()
```

### Archive Dataframe Cleaning

##### Define

Rows that are retweets need to be removed as they are not original. 

##### Code

```python
#Drop rows that have a retweeted status that isn't NaN.
archive_clean.drop(archive_clean[archive_clean['retweeted_status_id'].notnull()== True].index,inplace=True)
```

##### Test

```python
# all NaN values
archive_clean.info()
```

##### Define

Text variable is truncated and ideally all content should be visible.

##### Code

```python
pd.set_option('display.max_colwidth', -1)
```

##### Test

```python
archive_clean.head(1)
```

##### Define

Remove character reference for ampersand (&amp;) in the text variable and replace with just an ampersand symbol (&).

##### Code

```python
archive_clean['text'] = archive_clean['text'].str.replace('&amp;', '&')
```

##### Test

```python
archive_clean[archive_clean.text.str.contains('&amp;')]
```

##### Define

Convert source values from messy HTML anchor tag strings to tidy categories.

##### Code

```python
# Remove url from sources
archive_clean['source'] = archive_clean['source'].str.replace('<a href="http://twitter.com/download/iphone" rel="nofollow">Twitter for iPhone</a>', 'iPhone')
archive_clean['source'] = archive_clean['source'].str.replace('<a href="http://vine.co" rel="nofollow">Vine - Make a Scene</a>', 'Vine')
archive_clean['source'] = archive_clean['source'].str.replace('<a href="http://twitter.com" rel="nofollow">Twitter Web Client</a>', 'Web')
archive_clean['source'] = archive_clean['source'].str.replace('<a href="https://about.twitter.com/products/tweetdeck" rel="nofollow">TweetDeck</a>', 'TweetDeck')
```

```python
# Convert to categorical variable
archive_clean['source'] = archive_clean['source'].astype('category')
```

##### Test

```python
print(archive_clean['source'].value_counts())
print(archive_clean.info())
```

##### Define

Fix incorrect dog names by replacing them with the string value 'None".

##### Code

```python
#List is words that have been listed as dog names.
wrong_names = ['such', 'a', 'quite', 'not', 'one', 'incredibly', 'mad', 'an', 
               'very', 'just', 'my', 'his', 'actually', 'getting', 'the', 'light', 
               'life', 'infuriating', 'old', 'this', 'unacceptable', 'all', 'by', 
               'officially', 'space']

for name in wrong_names:
    archive_clean['name'].replace(name, 'None', inplace=True)
```

##### Test

```python
archive_clean[(archive_clean['name'].str.islower())]
```

##### Define

Convert dog names with values of 'None' into NaN's

##### Code

```python
#Converting names with value 'None' into NaN's
archive_clean['name'] = archive_clean['name'].replace('None', np.NaN)
```

##### Test

```python
archive_clean['name'].isnull().count()
```

##### Define

Enter the correct numerator and denominator for the entry that currently has a denominator of 0.

##### Code

```python
#Changing numerator and denominator to those listed in the text.
archive_clean.loc[(archive_clean['tweet_id'] == 835246439529840640), 'rating_numerator'] = 13
archive_clean.loc[(archive_clean['tweet_id'] == 835246439529840640), 'rating_denominator'] = 10
```

##### Test

```python
archive_clean[archive_clean.tweet_id == 835246439529840640]
```

##### Define

Fix dog stages variable by consolidating it into a single column titled 'stage'. Following this, change all None values to NaN's.

##### Code

```python
#Creating a stage variable and setting all values to 'None'
archive_clean['stage'] = "None"

# list of stages
stages = ['doggo', 'floofer', 'pupper', 'puppo']

#filling stage variable with those listed from the 4 columns
for stage in stages:
    archive_clean.loc[(archive_clean[stage] == stage), "stage"] = stage
```

```python
#Converting None values to NaN's
archive_clean['stage'] = archive_clean['stage'].replace('None', np.NaN)
```

##### Test

```python
archive_clean.info()
```

##### Define

Convert the timestamp column from a string into two separate datetime variables for date and time.

##### Code

```python
archive_clean['timestamp'] = pd.to_datetime(archive_clean['timestamp'])
```

```python
archive_clean['date'] = archive_clean['timestamp'].apply(lambda time: time.strftime('%d-%m-%Y'))
archive_clean['time'] = archive_clean['timestamp'].apply(lambda time: time.strftime('%H:%M:%S'))
```

##### Test

```python
archive_clean.head()
```

##### Define

Convert the tweet id variable from an integer to a string.

##### Code

```python
archive_clean['tweet_id'] = archive_clean['tweet_id'].astype(str)
```

##### Test

```python
archive_clean.info()
```

##### Define

Alter the numerator for those ratings that specifically have a float rating displayed in the 'text' section. We will need to firstly convert the numerator and denominator variables to floats and then programmatically update the numerator variables with the correct float value.

```python
# View tweets with float rating numerators in 'text' column
archive_clean[archive_clean['text'].str.contains(r"(\d+\.\d*\/\d+)")]
```

##### Code

```python
#Converting rating variables to floats
archive_clean['rating_numerator'] = archive_clean['rating_numerator'].astype('float')
archive_clean['rating_denominator'] = archive_clean['rating_denominator'].astype('float')
```

```python
archive_clean.loc[(archive_clean['tweet_id'] == '883482846933004288'), 'rating_numerator'] = 13.5
archive_clean.loc[(archive_clean['tweet_id'] == '786709082849828864'), 'rating_numerator'] = 9.75
archive_clean.loc[(archive_clean['tweet_id'] == '778027034220126208'), 'rating_numerator'] = 11.27
archive_clean.loc[(archive_clean['tweet_id'] == '681340665377193984'), 'rating_numerator'] = 9.5
archive_clean.loc[(archive_clean['tweet_id'] == '680494726643068929'), 'rating_numerator'] = 11.26
```

##### Test

```python
archive_clean[archive_clean['text'].str.contains(r"(\d+\.\d*\/\d+)")]
```

### Image Predictions Dataframe Cleaning

##### Define

Remove the underscores in the p1, p2 and p3 columns and replace with a space.

##### Code

```python
# List of columns to edit
pred_cols = ['p1', 'p2', 'p3']

# for loop of columns to replace underscores with spaces
for col in pred_cols:
    pred_clean[col] = pred_clean[col].str.replace('_', ' ')
```

##### Test

```python
pred_clean.head()
```

##### Define

Capitalise all the dog breeds in the p1, p2 and p3 columns.

##### Code

```python
# for loops of columns to capitalise words
for col in pred_cols:
    pred_clean[col] = pred_clean[col].str.title()
```

##### Test

```python
pred_clean.head()
```

Convert the tweet id variable from an integer to a string.

##### Code

```python
pred_clean['tweet_id'] = pred_clean['tweet_id'].astype(str)
```

### Dropping Unnecessary Data

##### Define

Drop unnecessary data from each of the dataframes.

##### Code

```python
#Dropping columns from archive dataframe
dropcols = ['in_reply_to_status_id', 'in_reply_to_user_id', 'timestamp', 'retweeted_status_id', 'retweeted_status_user_id', 
            'retweeted_status_timestamp', 'expanded_urls', 'doggo', 'floofer', 'pupper', 'puppo']
archive_clean = archive_clean.drop(dropcols, axis=1)
```

##### Test

```python
archive_clean.head()
```

### Merging

##### Define

Perform inner joins on dataframe to create a master dataframe.

##### Code

```python
df_master = pd.merge(archive_clean, info_clean, on='tweet_id', how='inner')
df_master = pd.merge(df_master, pred_clean, on='tweet_id', how='inner')
```

### Relationship between likes and retweets

```python
g = sns.JointGrid(x="likes", y="retweets", data=df_master, size=12)
g.plot_joint(sns.regplot, order=2)
g.plot_marginals(sns.distplot)
g.set_axis_labels("Likes", "Retweets",fontsize=13)
g.fig.suptitle("Relationships between likes and retweets", verticalalignment= 'top', size=15)
g.fig.subplots_adjust(top=.9);
```

Very evident positive relationship identified.
